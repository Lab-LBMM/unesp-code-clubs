1. Definir variáveis já fornecidas

```python
sequencia = 'ACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTATGCGTACGTTAGCTAGATTACAGATTACAGATTACAGATTACAGATTACAGATTACAGATTACAGATTACAGATTACAGATTACAGATTACAGAT'
primer = 'ATGCGTACGTTAGCTAGG'
```

2. Verificar se o primer está presente na sequência

```python
posicao = sequencia.find(primer)
```

3. Estrutura if/else para responder se o primer está na sequência e mostrar a posição

```python
if posicao != -1: #Isto é, se estiver presente
    print('O primer foi encontrado.')
    print(f'O primer foi encontrado na posição {posicao + 1}!')
else: 
    print('O primer não foi encontrado.')
```

4. Outra opção

```python
if primer in sequencia:
    print ('O primer está presente.')
else:
    print('O primer não está presente.')
```
